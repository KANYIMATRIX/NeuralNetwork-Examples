```python
import numpy as np
import glob
import skimage
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
```

```python
def extract(s):
    x = s.split("_x")[1].split("_y")[0]
    y = s.split("_x")[1].split("_y")[1].split("_class")[0]
    c = s.split("_x")[1].split("_y")[1].split("_class")[1].split(".")[0]
    return int(x),int(y), int(c)
```

```python
img = np.ones([2300,1750, 3])*254
imgc = np.ones([2300,1750, 3])*150
```

```python
for s in glob.glob("/Users/ieee8023/Movies/IDC_regular_ps50_idx5/8917/*/*"):
    x,y,c = extract(s)

    subimg = imread(s)
    img[y:y+subimg.shape[1], x:x+subimg.shape[0], :] = subimg.astype(int)
    imgc[y:y+subimg.shape[1], x:x+subimg.shape[0], :] = c*225
```

```python
plt.figure(figsize=(20,10))
plt.imshow(img.astype(int))
```

```python
plt.figure(figsize=(20,10))
plt.imshow(imgc.astype(int))
```
